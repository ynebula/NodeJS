```javascript
process.on('exit', function (code){
    console.log('exit event : ', code);
});

process.once('exit', function (code){
    console.log('exit event with once : ', code);
});

process.emit('exit');
process.emit('exit', 0);
process.emit('exit', 1);
```

```javascript
process.on('uncaughtException', function (code){
    console.log('uncaughtException');
});
```
